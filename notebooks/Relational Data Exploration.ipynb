```python
import sys
sys.path.append("../src/")
from data.readers import read_train_data
```

```python
features, labels = read_train_data("/home/mila/teaching/user09/data/Train")
```

```python
like_graph = dict()
```

```python
for user in features:
    for like in user.likes:
        if like not in like_graph:
            like_graph[like] = list()
        like_graph[like].append(user.user_id)
```

```python
list(like_graph.keys())[2]
```

```python
sorted_likes = sorted(
    list(like_graph.keys()),
    key=lambda x: len(like_graph.get(x)),
    reverse=True
)
```

```python
top_likes = sorted_likes[:20000]
```

```python
for like in top_likes:
    print(like, (len(like_graph.get(like)) / float(len(features))) * 100)
```
