```python
liwc_df = pd.read_csv(path_to_text + "liwc.csv")
nrc_df = pd.read_csv(path_to_text + "nrc.csv")
```

### Outlier detection with LIWC

```python
liwc_df.head(n=2)
```

```python
rng = np.random.RandomState(42)
clf = IsolationForest(max_samples=100, random_state=rng, behaviour="new", contamination=.1)
```

```python
X_train_liwc = liwc_df.values[:, 1:]
X_train_nrc = nrc_df.values[:, 1:]
```

```python
clf.fit(X_train_liwc)
y_pred_train = clf.predict(X_train_liwc)
```

```python
liwc_outliers = liwc_df[np.where(y_pred_train == -1, True, False)]
```

```python
print('There are {} outliers in the liwc dataset.'.format(len(liwc_outliers)))
liwc_outliers.head(n=5)
```

### Outlier detection with NRC

```python
clf.fit(X_train_nrc)
y_pred_train = clf.predict(X_train_nrc)
```

```python
nrc_outliers = nrc_df[np.where(y_pred_train == -1, True, False)]
```

```python
print('There are {} outliers in the nrc dataset.'.format(len(nrc_outliers)))
nrc_outliers.head(n=5)
```
